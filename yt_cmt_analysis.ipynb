```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from textblob import TextBlob
```

```python
comments=pd.read_csv('C:\\Users\\USER\\Desktop\\DSP/GBcomments.csv',error_bad_lines=False)
```

```python
comments.isna().sum()
```

```python
comments.dropna(inplace=True)
```

```python
comments['comment_text'][0]
```

```python
polarity=[]
for i in comments['comment_text']:
    polarity.append(TextBlob(i).sentiment.polarity)
```

```python
comments['polarity']=polarity
```

```python
comments_positive=comments[comments['polarity']==1]
```

```python
comments_positive.shape
```

```python
pip install wordcloud
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
stopwords=set(STOPWORDS)
```

```python
pos_words=" ".join(comments_positive['comment_text'])
```

```python
poswordcloud=WordCloud(width=1000,height=500,stopwords=stopwords).generate(pos_words)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(poswordcloud)
plt.axis('off')
```

```python
comments_negetive=comments[comments['polarity']==-1]
```

```python
neg_words=" ".join(comments_negetive['comment_text'])
```

```python
negwordcloud=WordCloud(height=500,width=1000,stopwords=stopwords).generate(neg_words)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(negwordcloud)
plt.axis('off')
```
